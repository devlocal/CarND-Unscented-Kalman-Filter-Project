```python
import matplotlib.pyplot as plt
```

```python
def plot_nis(file_name, trs, title):
    d= []
    with open(file_name) as f:
        d = [float(l) for l in f]

    fig = plt.figure(figsize=(20,10))
    fig.suptitle(title, fontsize=24, fontweight='bold')
    ax = fig.add_subplot(111)
    ax.set_xlabel('Measurements')
    ax.set_ylabel('NIS')
    plt.plot(d)
    plt.axhline(y=trs, color='darkred', linestyle='-')
    plt.show()
```

```python
plot_nis("lidar.txt", 5.991, "Lidar NIS")
```

```python
plot_nis("radar.txt", 7.815, "Radar NIS")
```
